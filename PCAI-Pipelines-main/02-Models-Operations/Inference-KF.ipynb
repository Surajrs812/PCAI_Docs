## Inference test

```python
!kubectl expose pod tf-serve --type=NodePort --port 8501
```

```python
%%capture cmd_output
!kubectl get svc | grep tf-serve | awk  '{print $3}'
```

```python
ipaddr = cmd_output.stdout[:-2]
print(f"Endpoint address: {ipaddr}")
```

```python
import os
import requests

# Repeat model_name in case Kubeflow cells were not execute manually 
model_name = "digitrec_model"

#skip proxy
os.environ['NO_PROXY'] = f"{ipaddr}"

#Build TF inference endpoint
inference_endpoint = f"http://{ipaddr}:8501/v1/models/{model_name}"
print(f"Inference endpoint: {inference_endpoint}")

# Build HTTP request
response = requests.get(inference_endpoint)

print("\nResponse:\n" + response.content.decode("utf-8"))
```
