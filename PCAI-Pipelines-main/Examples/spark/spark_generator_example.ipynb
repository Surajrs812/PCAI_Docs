## Prerequisite

```pyspark
%%spark
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, sum

# Define the file path for the CSV file
file_path = "/mounts/shared-volume/user/csv_spark/generator_sample_data.csv"

# Read the CSV file into a DataFrame
df = spark.read.csv(file_path, header=True, inferSchema=True)
cleaned_df = df.withColumn("Rental_Rate", col("Rental Rate (per day)").substr(1, 2).cast("int")).select("Model", "Location", "Rental_Rate")
 
# Group by Model and Location, and calculate the sum of rental rates
result_df = cleaned_df.groupBy("Model", "Location") \
                      .agg(sum("Rental_Rate").alias("Total_Rental_Rate"))
# Show the result
result_df.show() 
```
